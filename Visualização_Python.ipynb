# **Visualizando dados em Python**

## Comparando dados

### Gráfico de barras

Visualização utilizada para comparar dados dentro de grupos e entre grupos. 

Um grande desafio para aplicação deste tipo de visualização é a identificação do elemento que queremos apresentar. Como contadores de histórias com dados, devemos direcionar no público-alvo para os elementos que são mais relevantes:

- Qual elemento é o pivô da história que queremos contar?
- Existe algum elemento de nossa história que desejamos apresentar com mais detalhes ou chamar atenção?

O gráfico de barras é uma visualização fácil de compreender e interpretar pois se baseia na comparação simples de tamanhos de barras retangulares.

Podemos posicionar as barras verticalmente e horizontalmente, sendo esta última opção um recurso interessante quando temos rótulos de eixo muito grandes.

**Atenção! Devemos SEMPRE iniciar nossos eixos no zero, para não confundir o público-alvo e/ou direcioná-lo para uma interpretação errada dos dados.**

### Gráfico de barras clássico

```python
# Importando as bibliotecas
import matplotlib.pyplot as plt

# Criando dados de renda média por faixa etária para execução das visualizações
faixaEtaria = ['18-25', '26-35', '36-45', '46-55', '55+']
renda = [1500, 2130, 3540, 4110, 3280]

# O método 'bar' criará o gráfico de barras 
plt.bar(faixaEtaria, renda, color="green")
# Definindo as legendas de cada barra no eixo X
# O rótulo do eixo Y
plt.ylabel('Renda média')
# O rótulo do eixo X
plt.xlabel('Faixa etária (anos)')
# Um título para o gráfico
plt.title('Renda média por Faixa etária')
```

### Gráfico de barras horizontal

```python
# Importando as bibliotecas
import matplotlib.pyplot as plt

# Simulando a participação dos 5 produtos agrícolas produzidos no Brasil
# Definindo os produtos, que estarão no eixo X
produtos = ['Soja em grão', 'Cana de Açucar', 'Milho em grão', 'Café em grão', 'Mandioca']
# Definindo as participações que estarão no eixo Y
quantidade_pct = [33.6, 16.8, 10.3, 6.2, 3.8]

# Se fizéssemos o gráfico tradicional?
plt.bar(produtos, quantidade_pct, color="green")
```

```python
# Gráfico tradicional
plt.bar(produtos, quantidade_pct, color="green")

# Para resolver os rótulos do eixo X podemos tentar rotacioná-los
plt.xticks(rotation=90)
# Este método garante que a visualização fique ajustada à nossa janela
plt.tight_layout()
```

```python
# Vamos transformar em um gráfico de barras horizontal?
# No lugar do bar() usaremos o barh()
plt.barh(produtos, quantidade_pct, color='green')
plt.tight_layout()

# Finalizando com as informações
plt.ylabel('Produtos agrícolas')
plt.xlabel('Participação (em %)')
plt.title('Produtos com maior participação da indústria agrícola')
plt.tight_layout()
```

### Gráfico de barras empilhadas

```python
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Neste gráfico vamos representar a quantidade de ouro, pedra e madeira 
# de 3 cidades 
# Criando dados fictícios para cada um dos recursos, por cidade
ouro  = np.array((95, 80, 350))
pedra = np.array((110, 150, 230))
madeira = np.array((85, 325, 130))

# Gerando dados relacionados aos recursos de três cidades 
recursos = ['Ouro','Pedra','Madeira']
cidades = ['Cidade A','Cidade B','Cidade C']

# Definindo o tamanho da visualização
plt.figure(figsize=(10,5))

# Criando as barras
plt.bar(cidades, ouro, color='green')
plt.bar(cidades, pedra, color='blue', bottom=ouro)
plt.bar(cidades, madeira, color='red', bottom=ouro + pedra)

# Adiciando legendas as barras
plt.ylabel('Quantidade de recursos')
plt.title('Distribuição de recursos')
plt.legend(('Ouro','Pedra','Madeira'))
```

### Gráficos de área

Os gráficos de área nada mais são que gráficos de linha onde preenchemos toda a área abaixo da linha. 

Este tipo de visualização é muito útil quando queremos chamar a atenção para comportamentos diferenciados entre os dados.

```python
# Importando as bibliotecas
import matplotlib.pyplot as plt
import pandas as pd

# Criando dados de evolução de vendas no tempo, agora de três produtos diferentes
dados = {'ano': [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021],
         'unidades_vendidas_1': [532,603,674,444,1071,1970,2027,1169,1460,732,925],
         'unidades_vendidas_2': [812,801,1034,1101,1093,1231,987,1021,1302,1020,1103],
         'unidades_vendidas_3': [2025,2130,2143,1994,2087,2132,2027,1897,1901,1743,1830]
        }
df = pd.DataFrame(dados) 

# Definindo as cores que serão utilizadas
mapa_cores = ['red', 'steelblue', 'pink']
    
# Criando o gráfico de área mostrando as vendas acumuladas a cada ano
plt.stackplot(df['ano'], df['unidades_vendidas_1'],
              df['unidades_vendidas_2'],
              df['unidades_vendidas_3'],
              labels=['Produto 1', 'Produto 2', 'Produto 3'],
              colors=mapa_cores)

# Ajustando a visualização
plt.xlabel('Ano')
plt.ylabel('Vendas acumuladas')
plt.title('Vendas acumuladas por ano')
plt.legend()
```

### Gráfico de Pareto

O Gráfico de Pareto é um histograma ordenado e que apresenta, além das barras que quantificam os dados em cada caixa, um gráfico de linha agregado que representa o somatório da relevância de cada caixa no total. 

O Princípio de Pareto estabelece que “80% dos problemas são atribuídos a 20% das causas.”. Foi apresentado por Joseph M. Juran, que batizou este princípio com o nome do economista italiano Vilfredo Pareto, que observou que 80% da renda italiana era destinada para 20% da população (1897). É citada em exemplos extremamente comuns no dia a dia: "80% das nossas vendas vêm de 20% dos nossos clientes.”, sendo utilizada em diversas situações de processo de tomada de decisão.

#### Importando as bibliotecas

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter # Vamos usar isso na formatação numérica
import pandas as pd
```

Criando dados de evolução de vendas no tempo, agora de três produtos diferentes

```python
dados = {'produto': ['A','B','C','D','E','F','G','H','I','J','K'],
         'unidades_vendidas': [532,603,674,444,1071,1970,2027,1169,1460,732,925]
        }
df = pd.DataFrame(dados) 
df
```

Ordenando o DataFrame por contagem descendente

```python
df = df.sort_values(by='unidades_vendidas', ascending=False)
df.reset_index(drop=True, inplace=True)
df
```

Adicionando coluna com percentual acumulado de vendas

```python
df['perc_acum'] = df['unidades_vendidas'].cumsum()/df['unidades_vendidas'].sum()*100
df
```

Criação da visualização

```python
# Definições para a visualização
cor1 = 'steelblue'
cor2 = 'red'
esp_linha = 4

# A partir daqui é importante que todas as instruções sigam em uma mesma célula 
# de execução pois será construída em etapas.

# Criando o gráfico de barras base
fig, ax = plt.subplots()
ax.bar(df['produto'], df['unidades_vendidas'], color=cor1)
# Ajustando rótulos e títulos
ax.set_ylabel('Vendas por produto')
ax.set_xlabel('Produtos')
ax.set_title('Impacto da venda de produtos no total vendido')

# Adicionando linha com o percentual acumulado à visualização
# Criando um segundo eixo
ax2 = ax.twinx()
ax2.plot(df['produto'], df['perc_acum'], color=cor2, marker="D", ms=esp_linha)
# Usando um recurso do Matplotlib para apresentar percentuais da forma correta
ax2.yaxis.set_major_formatter(PercentFormatter())
# Incluindo um rótulo no eixo secundário
ax2.set_ylabel('Percentual acumulado')

# Especificando cores dos eixos
ax.tick_params(axis='y', colors=cor1)
ax2.tick_params(axis='y', colors=cor2)
```

## Visualizando relações

### Diagrama de dispersão

A visualização mais conhecida desta família de visualizações é o diagrama de dispersão. Nele conseguimos avaliar como duas variáveis se relacionam entre si e podemos perceber padrões de comportamento comuns que, apesar de muitas vezes não serem percepções definitivas, nos ajudam a melhorar nosso entendimento sobre a dinâmica do comportamento das variáveis.


O diagrama de dispersão é um gráfico chamado cartesiano, composto por dois eixos que, tradicionalmente, se cruzam onde os valores das duas variáveis são zero. Colocamos os dados de uma das variáveis ao longo do eixo X e os valores da outra variável em estudo no eixo Y. Finalmente, marcamos o cruzamento dos dados e avaliamos o resultado.


A relação que estamos buscando compreender com o diagrama de dispersão chama-se correlação. Uma correlação positiva indica que as duas variáveis se movem na mesma direção, e uma correlação negativa indica movimentação em direções opostas.


No diagrama de dispersão, visualizamos a correlação a partir da apresentação de uma reta de apoio chamada reta de tendência ou de regressão. Esta reta é criada a partir dos dados e apresenta o “comportamento médio” das duas variáveis. Caso a reta tenha um comportamento crescente, as duas variáveis têm uma correlação positiva e quanto maior a correlação, mais próximos os pontos estão da reta. No caso da reta ser decrescente, teremos uma correlação negativa e, novamente, quanto mais próximos da reta os pontos, mais negativa a correlação. Caso não tenhamos uma tendência clara, nem crescente, nem decrescente, então dizemos que não existe correlação entre as variáveis.

Devemos, porém, lembrar sempre que “correlação não indica causalidade”, ou seja, ter correlação positiva ou negativa alta não necessariamente quer dizer que as variáveis têm uma relação explicativa entre elas. **Esse é um dos enganos mais comuns cometidos em análise de dados!**

Uma outra visualização que pode ser usada para avaliar relações entre variáveis é o gráfico de bolhas, já visto na aula sobre comparação de dados.

```python
### Gráfico Radar

O gráfico radar é uma visualização onde apresentamos as informações em eixos que partem do centro em direção ao extremo do gráfico. Desta forma o resultado apresenta o dado com uma função de área e conseguimos comparar comportamentos entre dois grupos de dados diferentes.

Muito usado em jogos eletrônicos e RPG (Role-Playing Games) para indicar as características de um personagem ou atleta, tais como força, habilidade etc.
```

```python
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Criando os dados para a visualização
categories = ['Cardápio', 'Variedade', 'Serviço', 'Ambiente', 'Valor','Cardápio']
restaurante_1 = [4, 4, 5, 4, 3, 4]
restaurante_2 = [5, 5, 4, 5, 2, 5]
restaurante_3 = [3, 4, 5, 3, 5, 3]

# Criando a localizaçao das etiquetas em radianos, para posicionamento na visualização.
label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(restaurante_1))

# Criando a visualização
plt.figure(figsize=(8, 8))
# Necessário para o gráfico radar
plt.subplot(polar=True)
plt.plot(label_loc, restaurante_1, label='Restaurante 1')
plt.plot(label_loc, restaurante_2, label='Restaurante 2')
plt.plot(label_loc, restaurante_3, label='Restaurante 3')
plt.title('Comparação de restaurantes', size=20, y=1.05)
# Necessário para colocar os nomes das categorias nas posições corretas
lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories)
plt.legend()
```

## Visualizando partes do todo

### Gráfico de setores

Os gráficos de setores (popularmente conhecidos por gráficos de pizza) são, provavelmente, as visualizações mais conhecidas e mais utilizadas, principalmente na área de negócios de forma geral. No entanto, esta visualização apresenta problemas com relação à percepção correta da informação e, como cientistas de dados, devemos entender estas deficiências e utilizar a visualização da forma correta.

Nosso cérebro tem uma dificuldade em perceber diferenças baseadas em ângulos e este é o principal recurso utilizado em um gráfico de setores. Por isso, não é aconselhado utilizar uma visualização deste tipo para apresentar mais de duas ou três partes. Mais setores em um mesmo gráfico limita bastante nossa capacidade de identificar corretamente a representatividade das partes em relação ao todo.

A regra de ouro de um gráfico de setores é que o total deve sempre somar 100%. Caso deseje reduzir o número de setores, deve-se agregar as partes menores em uma única, deixando separadas as partes maiores. Uma outra boa prática é apresentar os setores em ordem crescente de representatividade ou, caso isso não faça sentido, em uma ordem lógica natural para os dados (por grupos de idades, por exemplo).

```python
# Importando as bibliotecas
import matplotlib.pyplot as plt
  
# Dados de vendas mensais por estado (Região Sudeste, em milhões de R$)
vendas = [44, 45, 31, 19]
estado = ['RJ', 'SP', 'MG', 'ES']
  
# Gerando a visualização com o Matplotlib
plt.pie(vendas, labels=estado, autopct='%.0f%%')

# Configurações gerais
plt.title('Vendas mensais na Região Sudeste (milhões de R$)', size=14)

plt.tight_layout()
```

```python
# Se fizéssemos o gráfico barras?
plt.bar(estado, vendas)

# Configurações gerais
plt.title('Vendas mensais na Região Sudeste (milhões de R$)', size=14)

plt.tight_layout()
```

### Treemap

Uma boa opção ao gráfico de setores é apresentar a informação em um gráfico de barras ou em um Treemap (Mapa de árvore).

Criado por Bem Schneiderman, da Universidade de Maryland, a visualização chamada Treemap divide um quadrado ou retângulo em setores, ilustrando algum grau de hierarquia ou relação de parte com o todo. 

O Treemap é uma visualização bem compacta, podendo incluir grupos e informações diferentes, além de hierarquias que podem ser identificadas, via rótulos, diretamente na visualização.

```python
# Importando as bibliotecas
import plotly.express as px
import pandas as pd
import plotly.io as pio

# Estabelecendo que o Plotly apresentará as visualizações no navegador
pio.renderers.default='colab'

# Gerando um dataframe com dados  de vendas por fornecedores e regiões
dados = {'fornecedores': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
         'areas': ['Tecnologia', 'Tecnologia', 'Finanças', 'Finanças',
                   'Tecnologia', 'Tecnologia', 'Finanças', 'Finanças'],
         'regioes': ['Sudeste', 'Sudeste', 'Sudeste', 'Sudeste',
                     'Sul', 'Sul', 'Sul', 'Sul'],
         'vendas': [1, 3, 2, 4, 2, 2, 1, 4]
        }
df = pd.DataFrame(dados)

# Criando uma coluna para termos um nó raiz para incluir todos os dados
df['todos'] = 'Todos' 

# Gerando a visualização
fig = px.treemap(df, path=['todos', 'regioes', 'areas', 'fornecedores'], values='vendas')
# Estabelecendo a cor da categoria raiz
fig.update_traces(root_color='lightgrey')
# Definindo título e posicionamento
fig.update_layout(title_text='Mapa de fornecedores', title_x=0.5)
```

### Sunburst

Uma outra visualização que nos ajuda a ver partes de um todo é o chamado Sunburst (Explosão solar). É uma visualização que lembra um gráfico de setores, porém com características de um Treemap e esteticamente mais bonita e que chama mais atenção.

O Sunburst apresenta hierarquia de informações de dentro para fora, com o nível mais alto sendo representado no anel interior da visualização e aumentando a granularidade da informação de dentro para fora.

```python
# Importando as bibliotecas
import plotly.express as px
import pandas as pd
import plotly.io as pio

# Estabelecendo que o Plotly apresentará as visualizações no navegador
pio.renderers.default='colab'

# Gerando um dataframe com dados de vendas por fornecedores e regiões
dados = {'fornecedores': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
         'areas': ['Tecnologia', 'Tecnologia', 'Finanças', 'Finanças',
                   'Tecnologia', 'Tecnologia', 'Finanças', 'Finanças'],
         'regioes': ['Sudeste', 'Sudeste', 'Sudeste', 'Sudeste',
                     'Sul', 'Sul', 'Sul', 'Sul'],
         'vendas': [1, 3, 2, 4, 2, 2, 1, 4]
        }
df = pd.DataFrame(dados)

fig = px.sunburst(df, path=['regioes', 'areas', 'fornecedores'], values='vendas')
# Definindo título e posicionamento
fig.update_layout(title_text='Mapa de fornecedores', title_x=0.5)
```

### Utilização de mapas em visualizações

Os mapas trazem para as visualizações uma imagem conhecida para as pessoas e podem agregar um senso de pertencimento importante em algumas apresentações.

Em algumas situações, sequem precisamos de um mapa real como base para a visualização, a simples densidade de pontos já traz uma imagem conhecida e que conseguimos relacionar a um país ou localidade específica.

Um uso comum é utilizar um mapa como base para um mapa de calor que trará a informação desejada.

```python
# Importando as bibliotecas
import plotly.express as px
import plotly.io as pio

# Estabelecendo que o Plotly apresentará as visualizações no navegador
pio.renderers.default='colab'

# Capturando dados socioeconômicos mundiais da Fundação Gapminder
# <https://pt.wikipedia.org/wiki/Funda%C3%A7%C3%A3o_Gapminder>
# Ano, 2007
df = px.data.gapminder().query("year==2007")
df
```

```python

# Gerando a visualização
fig = px.scatter_geo(df, locations="iso_alpha", color="continent",
                     hover_name="country", size="pop",
                     projection="natural earth")

# Configurações gerais
fig.update_layout(title_text='População mundial em 2007 (Fonte: Fundação Gapminder)',
                  title_x=0.5)
```

```python
# Gerando uma visualização animada

# Capturando dados socioeconômicos mundiais da Fundação Gapminder
df = px.data.gapminder()

# Gerando a visualização
fig = px.scatter_geo(df, locations="iso_alpha", color="continent",
                     hover_name="country", size="pop",
                     animation_frame="year",
                     projection="natural earth")

# Configurações gerais
fig.update_layout(title_text='População mundial entre 1952 e 2007 (Fonte: Fundação Gapminder)',
                  title_x=0.5)
```

### Nuvem de palavras

Quando queremos demonstrar a ocorrência e/ou relevância de palavras específicas em um conjunto de palavras podemos utilizar a visualização chamada Nuvem de palavras.

A Nuvem de palavras é um recurso de visualização muito utilizado pela mídia de forma geral nos dias de hoje pois é de simples entendimento e bastante informativa, quando utilizada corretamente.

Podemos agregar informações a uma nuvem de palavras por meio de cores.

```python
# importar os pacotes necessários
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS

# carregando base com resumos de análises de clientes Air B&B para imóveis no
# Rio de Janeiro
# Dados extraídos da base da Air B&B
resumos = pd.read_excel('Resumos_AirBB.xlsx')


palavras_resumos = " ".join(s for s in resumos['summary'])

# Criando uma lista de stopwords
# Stopwords são palavras (normalmente conectores e verbos) que não interessam 
# para nossa análise visual.
stopwords = set(STOPWORDS)
stopwords.update(['da', 'meu', 'em', 'você', 'de', 'ao', 'os', 'é', 'e', 'o', 'a'])

# Gerando a nuvem de palavras
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=800).generate(palavras_resumos)

# Apresentando a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.set_axis_off()
ax.imshow(wordcloud, interpolation='bilinear')
```

```python
# Utilizando uma imagem para uma nuvem de palavras diferente

# Caregando a imagem em um formato apropriado para aplicação na nuvem de palavras
rio_mascara = np.array(Image.open('Cristo.jpg'))

# Gerando a nuvem de palavras, agora usando a imagem
wordcloud = WordCloud(stopwords=stopwords,
                      background_color='white',
                      width=1000, height=1000, max_words=2000,
                      mask=rio_mascara, max_font_size=200,
                      min_font_size=1).generate(palavras_resumos)

# Apresentando a imagem final
fig, ax = plt.subplots(figsize=(10,10))
ax.set_axis_off()
ax.imshow(wordcloud, interpolation='bilinear')
```
